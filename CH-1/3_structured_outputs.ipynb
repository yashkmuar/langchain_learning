# ** Using Pydantic Models**

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    setup: str = Field(descrpition = "The setup for the joke")
    punchline: str = Field(description = "The punchline for th joke")

```

```python
obj = llm_schema(**{"setup":"setup", "punchline":"some punchline"})
obj
obj.setup
obj.punchline
```

```python
llm_structured_output = llm_openai.with_structured_output(llm_schema)

result = llm_structured_output.invoke("Tell me a joke")

print(type(result))
print(result.punchline)
```
