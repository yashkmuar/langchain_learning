# ** Using Pydantic Models**

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    setup: str = Field(descrpition = "The setup for the joke")
    punchline: str = Field(description = "The punchline for th joke")

```

```python
obj = llm_schema(**{"setup":"setup", "punchline":"some punchline"})
obj
obj.setup
obj.punchline
```

```python
llm_structured_output = llm_openai.with_structured_output(llm_schema)

result = llm_structured_output.invoke("Tell me a joke")

print(type(result))
print(result.punchline)
print(result)
```

**Using TypeDict**


```python
from typing import TypedDict

class llm_schema_td(TypedDict):
    setup: str
    punchline: str
```

```python
obj = llm_schema_td({"setup":"some setup", "punchline":"some punchline"})
obj
```

Here in TypedDict, if you gonna replace setup with some other words, its not fail, because it is assuming on typed check only.

In below I get the result_td as a dict object, not as a pydantic object.

```python
llm_structured_typed_dict = llm_openai.with_structured_output(llm_schema_td)

result_td = llm_structured_typed_dict.invoke("Tell me a joke")
result_td
```
