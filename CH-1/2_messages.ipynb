**Messages**

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

llm_openai = ChatOpenAI(model="gpt-5-mini",temperature=0.4)
my_messages=[
    SystemMessage(content="You are gen-z assistant who always answers in a fun way!!"),
    HumanMessage(content="Bro, tell me a fact about Thomas & Friends")
]

llm_openai.invoke(my_messages).content
```

# **Prompts**

Prompts are the messages that are sent to the LLM. Prompts are more user-friendly and dynamic in nature.

```python
import os
from langchain_core.messages import HumanMessage
from langchain_core.messages import SystemMessage
from dotenv import load_dotenv

load_dotenv()

messages=[
    SystemMessage(content="You are a gen-z assistant who always answers in a fun way!!"),
    HumanMessage(content="Bro, tell me a fun-fact about Ayurveda!!")
]

llm_openai.invoke(messages).content
```

**Prompts Templates**


```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
import os
from dotenv import load_dotenv

load_dotenv()

user_input = input("Enter a topic for fun fact: ")


dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic" : user_input})

llm_openai.invoke(ready_prompt).content


```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are a {tone} assistant"),
    ("user","Write a interesting fact about {topic}. Return the result in a key-value pair with key as 'fact' and value as interesting fact")
])

user_input = input("Enter a topic that you want a interesting fact about it: ")
user_tone = input("Enter a tone or way you want AI model interacts with you: ")

ready_prompt = prompt_template.invoke({"topic":user_input, "tone":user_tone})

llm_openai.invoke(ready_prompt.messages).content
```
