# **Chains in LangChain**

```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

if os.environ.get("OPENAI_API_KEY"):
    print("Bro, you got the key. Now, worlds is yours")
else:
    print("debugg yrr")

```

**FIRST CHAIN**

```python
#TASK -1 [Prompt]

from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant"),
    ("human","{input}")
])

```

```python
#TASK-2 [LLM]

from langchain_core.output_parsers import StrOutputParser

llm_openai = ChatOpenAI(model="gpt-5-mini",temperature=0)
```

```python
#TASK-3 [Str Parser]

str_parser = StrOutputParser()
```

### **Manual Invocation**

```python
template = prompt_template.invoke({"input":"What is the capital of Meghalaya?"})

res = llm_openai.invoke(template)

final_result = res.content
final_result
```

### **Chain Invocation**

```python
chain = prompt_template | llm_openai | str_parser

chain.invoke("What is the capital of Uttarakhand?")
```
